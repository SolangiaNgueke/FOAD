###Data préparation


```python
def load(dossier, label):
    images = []
    labels = []
    for fichier in os.listdir(dossier):
        if fichier.endswith(".jpg") or fichier.endswith(".png"):
            chemin_image = os.path.join(dossier, fichier)
            image = Image.open(chemin_image)
            if image is not None:
                image = image.resize((224, 224))
                image_array = np.array(image).astype(np.float32)
                image_array = image_array/ 255.0
                image_array = img_to_array(image)
                images.append(image_array)
                labels.append(label)

    return np.array(images), np.array(labels)
```

```python
parasitized_img, parasitized_labels = load(parasitized, label = 1)
uninfected_img, uninfected_labels = load(uninfected, label = 0)
```

```python
print("Number of uninfected cells", len(uninfected_img))
print("Number of parasitized cells", len(parasitized_img))
print("Number of uninfected label", len(uninfected_labels))
print("Number of parasitized label", len(parasitized_labels))
```

```python
images = parasitized_img + uninfected_img
labels = parasitized_labels + uninfected_labels
```

### Split de données en train et en test

```python
from sklearn.model_selection import train_test_split
image_train, image_test, label_train, label_test = train_test_split(images, labels, test_size=0.2, random_state=24)
```

### images préproccessing

```python
x_train_prepocessed = preprocess_input(image_train)
x_test_preprocessed = preprocess_input(image_test)
```

### Data augmentation

```python
data_augmentation_generator = ImageDataGenerator(
    rotation_range=45,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.3,
    shear_range=0.2,
    horizontal_flip=True,
    validation_split=0.15,
    fill_mode='nearest',
)
```

```python
train_generator = data_augmentation_generator.flow(x_train_prepocessed, label_train, batch_size=64, subset="training", seed=42)
val_generator = data_augmentation_generator.flow(x_train_prepocessed, label_train, batch_size=64, subset="validation", seed=42)
```

```python
augmented_images,_ = next(train_generator)
```

### Affichage des images augmentées

```python
plt.figure(figsize=(20, 14))
for i in range(10):
    ax = plt.subplot(1, 10, i + 1)
    plt.imshow(augmented_images[i])
    plt.axis("off")
plt.show()
```

### Callbacks

```python
early_stopping= keras.callbacks.EarlyStopping( monitor="val_loss", patience=5, restore_best_weights=True)
reduce_lr =keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.0001)
```

### Création du modèle

```python
def create_model():
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
        MaxPooling2D((2, 2)),
        Dropout(0.3),

        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Dropout(0.3),

         Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Dropout(0.3),

        Flatten(),

        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
model = create_model()
model.summary()
```

###Entrainement du modèle

```python
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=30,
    callbacks=[reduce_lr, early_stopping]
)
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.legend(['train loss', 'validation loss'])
plt.show()
```

```python
model.save("malaria_classification.h5")
```
